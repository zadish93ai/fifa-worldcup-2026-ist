# FIFA World Cup 2026 — Live IST Calendar

Generates `world_cup_live_ist.ics` from the full 104-match schedule in `data/world_cup_2026_schedule.json`.

- Stadium kickoff times are converted to **Indian Standard Time (IST)**
- **Completed** matches show live scores in the title (e.g. `🎯 Match 1: 🇲🇽 Mexico 2 - 1 South Africa 🇿🇦 (FT)`)
- **Scheduled** matches show team names only

Scores are fetched automatically from the [openfootball](https://github.com/openfootball/worldcup.json) public feed (and optionally [football-data.org](https://www.football-data.org/) if `FOOTBALL_DATA_API_KEY` is set). GitHub Actions runs this every 3 hours and commits updates back to the repo.

```python
from calendar_generator import build_event_title, stadium_time_to_ist
from live_sync import sync_and_generate

completed_scores, matches = sync_and_generate()

print(f"✅ Generated {len(matches)} events → world_cup_live_ist.ics")
print(f"   Completed: {len(completed_scores)}  |  Scheduled: {len(matches) - len(completed_scores)}")
print()
print("Sample events:")
for match in matches[:3] + matches[-2:]:
    title = build_event_title(match)
    ist = stadium_time_to_ist(match["date"], match["time"], match["timezone"])
    print(f"  • {title}")
    print(f"    {ist.strftime('%Y-%m-%d %H:%M %Z')}")
```
